```python
import matplotlib.pyplot as plt
from pyqpp import QCircuit, QEngine, QNoisyEngine, QubitDepolarizingNoise, gates
```

```python
def plot_stats(stats: dict[str, int], title: str) -> None:
    plt.figure(figsize=(6, 4))
    plt.bar(list(stats.keys()), list(stats.values()))
    plt.title(title)
    plt.xlabel('Outcome')
    plt.ylabel('Counts')
    plt.tight_layout()
    plt.show()
```

```python
qc1 = QCircuit(1, 1)
qc1.gate(gates.H, 0)
qc1.measure_all()
print(qc1)
```

```python
engine1 = QEngine(qc1)
engine1.execute(1000)
stats1 = engine1.get_stats()
plot_stats(stats1, 'Single-qubit Hadamard measurement')
```

```python
qc2 = QCircuit(2, 2)
qc2.gate(gates.H, 0)
qc2.CTRL(gates.X, 0, 1)
qc2.measure_all()
print(qc2)
```

```python
engine2 = QEngine(qc2)
engine2.execute(1000)
stats2 = engine2.get_stats()
plot_stats(stats2, 'Bell-state measurement')
```

```python
noisy_engine = QNoisyEngine(qc2, QubitDepolarizingNoise(0.01))
noisy_engine.execute(1000)
noisy_stats = noisy_engine.get_stats()
plot_stats(noisy_stats, 'Bell-state measurement with depolarizing noise (p = 0.01)')
```
